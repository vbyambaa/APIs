```python
def get_weather(city):
    url = f'http://api.weatherapi.com/v1/current.json?key={byambaa_key}&q={city}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        current = data['current']
        location = data['location']
        temp_f = current['temp_f']
        wind_mph = current['wind_mph']
        humidity = current['humidity']

        print(f'Weather: {location ['name']}, {location['region']}')
        print(f'Temperature:{current['temp_f']} Fh / {current['temp_c']} C')
        print(f'Condition:{current['condition']['text']}')
        print(f'Humidity:{current['humidity']} %')
        print(f'Wind Speed:{current['wind_mph']} mph')

    else:
        print(f'Error Fetching Data for {city}: {response.status_code}')
```

```python
cities = ['Easton', 'Ulaanbaatar', 'Louisville', 'Santa Fe', 'Tosontsengel']
weather_data = []

for city in cities:
    url = f'http://api.weatherapi.com/v1/current.json?key={byambaa_key}&q={city}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        weather_data.append({
            'City' : data['location']['name'],
            'Temperature': data['current']['temp_f'],
            'Condition' : data['current']['condition'],
            'Humidity' : data['current']['humidity'],
            'Wind Speed' : data['current']['wind_mph']
        })
    else:
        print('Error fetching data')
```

```python
weather_df = pd.DataFrame(weather_data)
```
