```python
forecast_url = f"{baseurl}/forecast.json?key={bv_key}&q={city}&days=3"
forecast_response = requests.get(forecast_url)
forecast_response.status_code
```

```python
print(forecast_url)
```

```python
forecast_data = forecast_response.json()
forecast = forecast_data['forecast']['forecastday']
```

```python
forecast_3_days_summary = []
for day in forecast:
    forecast_3_days_summary.append(f'{day['date']}:high{day['day']['maxtemp_f']}F, low{day['day']['mintemp_f']} - {day['day']['condition']['text']}')
```

```python
forecast_3_days_summary
```

```python
forecast_3_days_summary = []
for day in forecast:
    date_object = datetime.strptime(day['date'], '%Y-%m-%d')
    formated_date = date_object.strftime("%A %B %#d")
    forecast_3_days_summary.append(
        f'{formated_date}: high {day['day']['maxtemp_f']}F / low {day['day']['mintemp_f']}F - {day['day']['condition']['text']}'
        )

print("\n".join(forecast_3_days_summary))
```

```python
today = pd.Timestamp.now().strftime('%Y-%m-%d')
astronomy_url = f"{baseurl}astronomy.json?key={bv_key}&q={city}&days={today}"
```

```python
def get_city_data(city, bv_key):
    baseurl = 'http://api.weatherapi.com/v1/'

    current_url = f'{baseurl}/current.json?key={bv_key}&q={city}'
    forecast_url = f"{baseurl}/forecast.json?key={bv_key}&q={city}&days=3"
    astronomy_url = f"{baseurl}astronomy.json?key={bv_key}&q={city}&days={today}"



    current_data = requests.get(current_url).json()
    forecast_data = requests.get(forecast_url).json()
    astronomy_data = requests.get(astronomy_url).json()

    
    return {
        'current' : current_data,
        'forecast' : forecast_data,
        'astronomy' : astronomy_data
    }

```

```python
data = get_city_data('Louisville', bv_key)
print(data['current'])
# print(data['forecast'])
# print(data['astronomy'])
```
